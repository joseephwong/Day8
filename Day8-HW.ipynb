**I am now happy with how the data is formatted. Although summary statistics for 'Season' are not necessary useful, having it be an int will allow me to use it as a unit of time and order/group chronologically.**

```python
# Make some sort of graph using columns in your data.
# Let's first observe if there is a relationship between times nominated and competition wins.
my_columns = ['Season','Houseguest','Times  Nominated','HoH &  PoV Wins']
fig = px.scatter(DF_raw[my_columns],
                   x = 'Times  Nominated',
                   y = 'HoH &  PoV Wins',
                   color_discrete_sequence = ['skyblue'],
                   hover_data = 'Houseguest')
fig.update_layout(bargap = 0.1,
                 title = 'Competition Wins vs. Times Nominated',
                 title_x = 0.5)
fig
```

**Competition Wins vs. Times Nominated Analysis**

The scatterplot suggests that there isn't an apparent relationship between number of times nominated and competition wins. There appears to be a cluster of data points between 2 and 4 times nominated. It makes sense that there would be some observations with 0 times nominated because there have been finalists who were never nominated but won competitions. With more data points, it could be argued that the distribution of wins over times nominated is a bell curve shape, peaking around 3 times nominated and 7 wins.

```python
# Next, let's explore how competition win percentages have changed over time (over the seasons).
my_columns = ['Season','Houseguest','Win %']
fig = px.scatter(DF_raw[my_columns],
                   x = 'Season',
                   y = 'Win %',
                   color_discrete_sequence = ['skyblue'],
                   hover_data = 'Houseguest')
fig.update_layout(bargap = 0.1,
                 title = 'Win % over the Seasons',
                 title_x = 0.5)
fig
```

**Win % over the Seasons Analysis**

The scatterplot suggests that there may be a slight increase in win percentage over time (in newer seasons). The maximum win percentage from each season appears to be gradually increasing, starting at 23.08% in Season 3, reaching a peak of 40% in Season 14 (excluding the outlier, Janelle Pierzina in Season 7). However, the overall shaping of the data is not drastic, so any potential relationship between win percentage and progression of time would be relatively weak.

Season 10 is interesting in that it only has 4 data points, with it's minimum (16.67%) being surprisingly high compared to the other seasons. Additionally, we should note the outlier of Janelle Pierzina in Season 7 with a 60% win rate. This graph really emphasizes just how impressive her statistic is; she won MORE than every other competition.
