```python
# Make a histogram of the acres burned and color the bars by whether or not the incident was a calfire incident.
# You will probably need to make a mask to remove very small and very large fires.
my_columns = ['incident_acres_burned','calfire_incident']
mask = (DF_raw['incident_acres_burned'] < 300000) & (DF_raw['incident_acres_burned'] > 5000)
DF_masked = DF_raw[mask]
fig = px.histogram(DF_masked[my_columns],
                   x = 'incident_acres_burned',
                   color = 'calfire_incident')
fig.update_layout(title = 'Count of acres burned (5,000 to 300,000 acres)',
                 xaxis_title = 'Acres Burned',
                 yaxis_title = 'Count')
fig
```

```python
# How many fires burned more than 100,000 acres?
mask = DF_raw['incident_acres_burned'] > 100000
DF_masked = DF_raw[mask]
len(DF_masked)
```

```python
# See if you can create a graph that answers the question:
# Are fires getting bigger or more frequent over time?
my_cols = ['incident_dateonly_created','incident_acres_burned']
fig = px.scatter(DF_raw[my_cols],
                 x = 'incident_dateonly_created',
                 y = 'incident_acres_burned')
fig.update_layout(title = 'Fire Size (acres) Over Time',
                  title_x = 0.5,
                  xaxis_title = 'Creation Date of Incident',
                  yaxis_title = 'Acres Burned')
fig
```

The scatterplot suggests that fire are not getting significantly bigger over time.

```python
my_cols = ['incident_dateonly_created','incident_acres_burned']
fig = px.histogram(DF_raw[my_cols],
                  x = 'incident_dateonly_created',
                  nbins = 100)
fig.update_layout(title = 'Count of Fires Over time',
                  title_x = 0.5,
                  xaxis_title = 'Creation Date of Incident',
                  yaxis_title = 'Count')
fig
```

The histogram suggets that the frequency of fires could potentially be increasing over time. However, the yearly number of fires fluxuates and does not have a steady increase. Instead, it the count increases, then decreases, then increases by more.
